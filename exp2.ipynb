```python
img2=cv2.imread('nature_20.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img2)
```

```python
valuhist=cv2.calcHist([img2],channels=[0],mask=None,histSize=[256],ranges=[0,256])
```

```python
plt.plot(valuhist)
```

```python
plt.imshow(img2,cmap='gray')


```

```python
hist_img3=cv2.calcHist([img2],channels=[0],mask=None,histSize=[256],ranges=[0,256])
plt.plot(hist_img3)
```

```python
equalize=cv2.equalizeHist(img2)
```

```python
plt.imshow(equalize,cmap='gray')
```

```python
hist_img4=cv2.calcHist([equalize],channels=[0],mask=None,histSize=[256],ranges=[0,256])
plt.plot(hist_img4)
```

```python
#fig=plt.figure(dpi=250)

plt.subplot(221),plt.imshow(img2,cmap='gray'),plt.title('Original Image')
plt.subplot(222),plt.imshow(equalize,cmap='gray'),plt.title('Equalized Image')
plt.subplot(223),plt.plot(hist_img3),plt.title('Original Plot')
plt.subplot(224),plt.plot(hist_img4),plt.title('Equalized plot')
```

```python
nature=cv2.imread('nature_20.jpg')
showNature=cv2.cvtColor(nature,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(showNature)
```

```python
img5=nature
color=('b','g','r')
for i, col in enumerate(color):
  histr=cv2.calcHist([img5],[i],None,[256],[0,256])
  plt.plot(histr,color=col)
  plt.xlim([0,256])
plt.title('hihi')  
```

```python
img6=nature
img6.shape
```

```python
mask=np.zeros(img6.shape[:2],np.uint8)
mask[20:50,80:120]=255
```

```python
plt.imshow(mask,cmap='gray')
```

```python
masked_img=cv2.bitwise_and(img6,img6,mask=mask)
show_masked_img=cv2.bitwise_and(showNature,showNature,mask=mask)
```

```python
plt.imshow(show_masked_img)
```

```python
hist_img6=cv2.calcHist([masked_img],channels=[0],mask=None,histSize=[256],ranges=[0,256])
plt.plot(hist_img6)
```

```python
fig=plt.figure(dpi=250)

plt.subplot(221),plt.imshow(img),plt.title('Original Image')
plt.subplot(222),plt.imshow(masked_img),plt.title('masked Image')
plt.subplot(223),plt.plot(histr,color=col),plt.title('Original Plot')
plt.subplot(224),plt.plot(hist_img6),plt.title('masked plot')
```
