```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
img_gray = cv2.imread('mona.jpg', cv2.IMREAD_GRAYSCALE)
# img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# plt.imshow(img)
plt.imshow(img_gray, cmap = 'gray')
```

```python
print(img_gray.shape)
```

```python
gauss_noise = np.zeros((188, 143), dtype = np.uint8)
cv2.randn(gauss_noise,128, 20)
gauss_noise = (gauss_noise*0.5).astype(np.uint8)
plt.imshow(gauss_noise, cmap = 'gray')
```

```python
gn_img = cv2.add(img_gray, gauss_noise)
plt.imshow(gn_img)
```

```python
fig = plt.figure(dpi = 150)

fig.add_subplot(1,3,1)
plt.imshow(img_gray, cmap = 'gray')
plt.axis('off')
plt.title('original')

fig.add_subplot(1,3,2)
plt.imshow(gauss_noise, cmap = 'gray')
plt.axis('off')
plt.title('Gaussian Noise')

fig.add_subplot(1,3,3)
plt.imshow(gn_img, cmap = 'gray')
plt.axis('off')
plt.title('Combined')

```

```python
#impulse Noise(salt and pepper)

img_noise = np.zeros((188, 143), dtype = np.uint8)
cv2.randn(img_noise, 0,255)
img_noise = cv2.threshold(img_noise, 245, 255, cv2.THRESH_BINARY)[1]
```

```python
in_img = cv2.add(img_gray, img_noise)
```

```python
fig = plt.figure(dpi = 200)

fig.add_subplot(1,3,1)
plt.imshow(img_gray, cmap = 'gray')
plt.axis('off')
plt.title('Original')

fig.add_subplot(1,3,2)
plt.imshow(img_noise, cmap = 'gray')
plt.axis('off')
plt.title('Impulse Noise')


fig.add_subplot(1,3,3)
plt.imshow(in_img, cmap = 'gray')
plt.axis('off')
plt.title('Combined')
```

```python
#blur image

blur3 = cv2.blur(in_img, (3,3))
blur9 = cv2.blur(in_img, (9,9))
blur25 = cv2.blur(in_img, (25,25))

fig = plt.figure(dpi = 200)
plt.subplot(141), plt.imshow(in_img, cmap = 'gray'), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(142), plt.imshow(blur3, cmap = 'gray'), plt.title('Blurred3')
plt.xticks([]), plt.yticks([])
plt.subplot(143), plt.imshow(blur9, cmap = 'gray'), plt.title('Blurred9')
plt.xticks([]), plt.yticks([])
plt.subplot(144), plt.imshow(blur25, cmap = 'gray'), plt.title('Blurred25')
plt.xticks([]), plt.yticks([])
```

```python
median3 = cv2.medianBlur(in_img, 3)
#plt.imshow(median3, cmap = 'gray')

median9 = cv2.medianBlur(in_img, 9)
#plt.imshow(median9, cmap = 'gray')

fig = plt.figure(dpi = 200)

plt.subplot(141), plt.imshow(in_img, cmap = 'gray'), plt.title('original')
plt.xticks([]), plt.yticks([])
plt.subplot(142), plt.imshow(median3, cmap = 'gray'), plt.title('Meadian3')
plt.xticks([]), plt.yticks([])
plt.subplot(143), plt.imshow(median9, cmap = 'gray'), plt.title('Median9')
plt.xticks([]), plt.yticks([])
```

```python
# Importing Image and ImageFilter module from PIL package
from PIL import Image, ImageFilter

# creating a image object
im1 = Image.open("mona.jpg")

# applying the max filter
im3 = im1.filter(ImageFilter.MaxFilter(size = 3))

plt.imshow(im3)

```
